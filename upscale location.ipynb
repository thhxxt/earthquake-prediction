```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

def upscale(k,file_path,t,iteration=1000):
    data = pd.read_csv(file_path)
    location = data[['Latitude','Longitude']]
    
    model = KMeans(n_clusters = k, max_iter = iteration)
    model.fit(location)
    r1 = pd.DataFrame(model.cluster_centers_) #找出聚类中心
    r = pd.concat([data, pd.Series(model.labels_, index = location.index)], axis = 1)  #详
    m = r.rename(columns = {0:'label'})
    
    for i in range(k):
        r.loc[r[0]==i,'Latitude'] = r1.at[i,0]
        r.loc[r[0]==i,'Longitude'] = r1.at[i,1]
        
    del r['Well #']
    
    if t == 0:
        f = r.rename(columns = {0:'label'}).groupby(['Production Date','Longitude','Latitude','label']).agg('sum').reset_index()
        return r1,m,f
    if t == 1:
        f = r.rename(columns = {0:'label'}).groupby(['Injection Date','Longitude','Latitude','label']).agg('sum').reset_index()
        return r1,m,f
    return 

def location_plot(data_production):
    fig = plt.figure(figsize=(10,10))
    df2 =pd.read_csv("all catalogs.csv")
    ax = fig.add_axes([0,0,1,1])
    y2=df2['Latitude']
    x2=df2['Longitude']
    ax.grid(color='g', ls = '-.') 
    ax.scatter(x2,y2,color='b', label='earthquake') 
    ax.scatter(data_production.loc[:,1],data_production.loc[:,0],color='r')
    plt.legend()
    plt.show()
    
def location_plot2(data_production,data_injection):
    fig = plt.figure(figsize=(10,10))
    df2 =pd.read_csv("all catalogs.csv")
    ax = fig.add_axes([0,0,1,1])
    y2=df2['Latitude']
    x2=df2['Longitude']
    ax.grid(color='g', ls = '-.') 
    ax.scatter(x2,y2,color='b', label='earthquake') 
    ax.scatter(data_production.loc[:,1],data_production.loc[:,0],color='r',label='wells production')
    ax.scatter(data_injection.loc[:,1],data_injection.loc[:,0],color='y',label='wells injection')
    plt.legend()
    plt.show()
    
def main():
    path_pro = './production_020202.csv'
    path_inj = './injection_020202.csv'
    p1,p2,p3 = upscale(100,path_pro,0)
    i1,i2,i3 = upscale(100,path_inj,1)
    location_plot(p1)
    location_plot(i1)
    location_plot2(p1,i1)
main()z
```
